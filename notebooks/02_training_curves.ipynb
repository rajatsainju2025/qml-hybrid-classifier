# Notebook 02 — Training Curves & Convergence Analysis

Reproduces the paper-style training plots from the main experiment.
Run `experiments/run_experiment.py` first to generate `results/` artefacts,
or run the training directly in this notebook.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent / "src"))

import matplotlib.pyplot as plt
import numpy as np

from qml_hybrid.train import run_training
from qml_hybrid.utils import plot_training_curves

# Config matching baseline_vqc.yaml
config = {
    "experiment_name": "notebook-training",
    "run_name": "nb02-strongly-entangling",
    "dataset": "breast_cancer",
    "n_qubits": 8,
    "n_layers": 3,
    "ansatz": "strongly_entangling",
    "embedding": "angle",
    "lr": 0.01,
    "batch_size": 32,
    "max_epochs": 100,  # Reduced for notebook speed; use 150 for full run
    "patience": 25,
    "seed": 42,
    "n_classes": 2,
    "val_size": 0.15,
    "test_size": 0.15,
    "results_dir": "../results",
}

print("Training VQC (this takes ~5-15 min on CPU for 100 epochs)...")
result = run_training(config)
print(f"\nTraining complete in {result['train_time_s']:.1f}s")
print(f"Stopped at epoch {len(result['train_losses'])}")
```

```python
fig = plot_training_curves(
    result["train_losses"],
    result["val_losses"],
    title="VQC Training: StronglyEntangling, 8 qubits, 3 layers",
)
plt.show()

# Quick test metrics
from qml_hybrid.evaluate import compute_metrics
from qml_hybrid.utils import load_dataset

_, _, _, _, _, y_test = load_dataset(seed=42)
metrics = compute_metrics(y_test, result["test_preds"], result["test_probs"])
print("\nTest metrics:")
for k, v in metrics.items():
    print(f"  {k:15s}: {v:.4f}")
```
